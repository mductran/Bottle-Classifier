```python
"""
bottle classifier
"""

import tensorflow as tf
import pickle
import os
import numpy as np
import cv2
from sklearn.utils import shuffle
from tensorflow.contrib.layers import flatten
import matplotlib.pyplot as plt


os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
```

```python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include
### converting to grayscale, etc.
def greyscale(img):
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

def normalize(img):
    img = np.divide(img-128, 128)
    return img

# shuffle data
```

```python
def LeNet(x):    
    # Hyperparameters
    mu = 0
    sigma = 0.1    
    
    # TODO: Layer 1: Convolutional. Input = 96x96x1. Output = 92x92x6.
    W1 = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean = mu, stddev = sigma))
    conv1 = tf.nn.conv2d(x, W1, strides=[1, 1, 1, 1], padding='VALID')
    b1 = tf.Variable(tf.zeros(6))
    conv1 = tf.nn.bias_add(conv1, b1)
    print("layer 1 shape:",conv1.get_shape())

    # TODO: Activation.
    conv1 = tf.nn.relu(conv1)
    
    # TODO: Pooling. Input = 92x92x6. Output = 46x46x6.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # TODO: Layer 2: Convolutional. Input = 46x46x6. Output = 40x40x16.
    W2 = tf.Variable(tf.truncated_normal(shape=(7, 7, 6, 16), mean = mu, stddev = sigma))
    conv2 = tf.nn.conv2d(conv1, W2, strides=[1, 1, 1, 1], padding='VALID')
    b2 = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.bias_add(conv2, b2)
    print("layer 2 shape:",conv2.get_shape())
                    
    # TODO: Activation.
    conv2 = tf.nn.relu(conv2)

    # TODO: Pooling. Input = 40x40x16. Output = 20x20x16.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # Another convolution. Input = 20x20x16. Output = 16x16x16
    W3 = tf.Variable(tf.truncated_normal(shape=(5, 5, 16, 16), mean = mu, stddev = sigma))
    conv3 = tf.nn.conv2d(conv2, W3, strides=[1, 1, 1, 1], padding='VALID')
    b3 = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.bias_add(conv3, b3)
    print("layer 3 shape:",conv2.get_shape())
    
    # One more pooling layer. Input = 16x16x16. Output = 8x8x16
    conv3 = tf.nn.max_pool(conv3, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # TODO: Flatten. Input = 8x8x16. Output = 1024.
    conv3 = flatten(conv3)
    
    # TODO: Layer 3: Fully Connected. Input = 1024. Output = 512.
    W4 = tf.Variable(tf.truncated_normal(shape=(1024, 512), mean = mu, stddev = sigma))
    b4 = tf.Variable(tf.zeros(512))    
    full1 = tf.add(tf.matmul(conv3, W4), b4)
    
    # TODO: Activation.
    full1 = tf.nn.relu(full1)
    
    # Dropout
    # full1 = tf.nn.dropout(full1, keep_prob)

    # TODO: Layer 4: Fully Connected. Input = 512. Output = 128.
    W5 = tf.Variable(tf.truncated_normal(shape=(512, 128), mean = mu, stddev = sigma))
    b5 = tf.Variable(tf.zeros(128)) 
    full2 = tf.add(tf.matmul(full1, W5), b5)
    
    # TODO: Activation.
    full2 = tf.nn.relu(full2)
    
    # Dropout
    # full2 = tf.nn.dropout(full2, keep_prob)

    # TODO: Layer 5: Fully Connected. Input = 128. Output = 32.
    W6 = tf.Variable(tf.truncated_normal(shape=(128, 32), mean = mu, stddev = sigma))
    b6 = tf.Variable(tf.zeros(32)) 
    full3 = tf.add(tf.matmul(full2, W6), b6)
    
    # One final FC layer. Input = 32, output = 4
    W7 = tf.Variable(tf.truncated_normal(shape=(32, 4), mean = mu, stddev = sigma))
    b7 = tf.Variable(tf.zeros(4)) 
    logits = tf.add(tf.matmul(full3, W7), b7)
    
    return logits

print('done')
```
