## Create training data : X_train and y_train  ( ! specify image_size )

```python
# Folder contains image and label file.
path = '/home/dvh/Downloads/realsense-d435-automated-master/training_data/'
csv_file = 'label.csv'

data = pd.read_csv(path+csv_file)
X_train = []
y_train = []

# resize image to a predefined size
image_size = (96,96)

for i in range(len(data)):   
    filename = data.iloc[i,0]
    label = data.iloc[i,1]

    img = plt.imread(path + filename) 
    img = cv2.resize(img, image_size, interpolation=cv2.INTER_CUBIC)
    X_train.append(img)        
    y_train.append(label)        
        
```

## Plot X_train and print y_train

```python
plt.figure(figsize=(30,15))
i = 0
for img in X_train:
    i+=1
    plt.subplot(1,len(X_train),i)
    plt.imshow(img)        
                    
plt.show()

print(y_train)
```
